```python
import pandas as pd
```

```python
movies = pd.read_csv("smaller-movie/movies.csv")
movies
```

```python
tmdb_movies = pd.read_csv("tmdb-5000-movie-dataset/tmdb_5000_movies.csv")
# tmdb_movies.index.name = "movieId"
tmdb_movies.sort_index(inplace=True)
release_year = tmdb_movies["release_date"].apply(lambda x: f" ({str(x)[:4]})" if len(str(x)) > 5 else None)
tmdb_movies["title"] = tmdb_movies["title"] + release_year
tmdb_movies
```

```python
pd.merge(movies, tmdb_movies, on= 'title')
```
